#### Task 1 - Regression Exercise

```python
df = pd.read_csv('fake_reg.csv')
df.head() #predict price based on feature1 and feature2 - supervised
```

```python
sns.pairplot(df);
```

```python
### Preparing the Data
X = df[['feature1', 'feature2']].values
y = df['price'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = MinMaxScaler()
scaler.fit(X_train) #calculate parameters - only use training set - prevent data leakage
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
### Creating and Training the Model
# first method to build the model
# model = Sequential([Dense(4, activation='relu'), #first layer - 4 neurons, densely connected
#                    Dense(2, activation='relu')，
#                    Dense(1)]) #output layer
# second method to build the model - preferred
model = Sequential()
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1)) #predict - single numerical value
model.compile(optimizer='rmsprop', loss='mse') #regression problem
```

```python
model.fit(x=X_train, y=y_train, epochs=250) #the model is going to go through the training set 250 times
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
```

```python
### Model Evaluation
model.evaluate(X_test, y_test, verbose=0) #verbose=0: no output #mse on test set
```

```python
model.evaluate(X_train, y_train, verbose=0) #mse on training set
```

```python
test_predictions = model.predict(X_test) #predicted results
test_predictions = pd.Series(test_predictions.reshape(300,))
pred_df = pd.DataFrame(y_test, columns=['Test True Y'])
pred_df = pd.concat([pred_df, test_predictions], axis=1)
pred_df.columns = ['Test True Y', 'Model Predictions']
pred_df
```

```python
sns.scatterplot(x = 'Test True Y', y = 'Model Predictions', data = pred_df); #straight line - good performance
```

```python
print(mean_absolute_error(pred_df['Test True Y'], pred_df['Model Predictions'])) #4$ differences
print(df['price'].describe()) #mae is small compared to actual mean of 'price' - well-performed model
print(mean_squared_error(pred_df['Test True Y'], pred_df['Model Predictions']))
print(mean_squared_error(pred_df['Test True Y'], pred_df['Model Predictions']) ** 0.5) #root mean squared error
```

```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(30, activation='relu')) #30 features
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid')) #binary classification
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot(); #over fitting -> early stopping
```

```python
model = Sequential()
model.add(Dense(30, activation='relu')) #30 features
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid')) #binary classification
model.compile(loss='binary_crossentropy', optimizer='adam')
early_stop = EarlyStopping(monitor='val_loss', mode='min', #minimize val_loss
                          verbose=1, patience=25) #wait 25 epochs after stopping point
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot();
```

```python
# another way to prevent over fitting - dropout
model = Sequential()
model.add(Dense(30, activation='relu')) #30 features
model.add(Dropout(0.5)) #turn off 50% of neurons (random) when training each batch
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid')) #binary classification
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot(); #better than previous one
```

```python
predictions = (model.predict(X_test) > 0.5).astype('int32') #binary classification
# preds_prob = model.predict(X_test) 
# predictions = np.argmax(preds_prob, axis=1) #for multi-class classification
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

## Section 10: Natural Language Processing

```python
# step 1: read in text data
# step 2: text processing and vectorization
# step 3: create batches
# step 4: model - embedding, GRU, dense
# step 5: train the model
# step 6: generate new text
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Embedding,Dropout,GRU
from tensorflow.keras.losses import sparse_categorical_crossentropy
```

```python
# Data
path_to_file = 'shakespeare.txt'
text = open(path_to_file, 'r').read()
print(text[:500])
vocab = sorted(set(text)) #all the distinct characters
```

```python
# Text Processing
char_to_ind = {char:ind for ind, char in enumerate(vocab)} #dictionary
print(char_to_ind['H'])
ind_to_char = np.array(vocab)
print(ind_to_char[33])
encoded_text = np.array([char_to_ind[c] for c in text]) #assign an unique integer to each character
encoded_text
```

```python
# Creating Batches
seq_len = 120
total_num_seq = len(text) // (seq_len + 1)
char_dataset = tf.data.Dataset.from_tensor_slices(encoded_text)
sequences = char_dataset.batch(seq_len+1, drop_remainder=True)

def create_seq_targets(seq): #seq = 'Hello my name'
    input_text = seq[:-1] #'Hello my nam'
    target_text = seq[1:] #'ello my name'
    return input_text, target_text

dataset = sequences.map(create_seq_targets)

# print the first batch
for input_txt, target_txt in  dataset.take(1):
    print(input_txt.numpy())
    print(''.join(ind_to_char[input_txt.numpy()]))
    print('\n')
    print(target_txt.numpy())
    print(''.join(ind_to_char[target_txt.numpy()]))
    
# generate training batches
batch_size = 128
buffer_size = 10000
dataset = dataset.shuffle(buffer_size).batch(batch_size, drop_remainder=True)
```

```python
# Creating the Model
vocab_size = len(vocab)
embed_dim = 64
rnn_neurons = 1026

def sparse_cat_loss(y_true,y_pred):
    return sparse_categorical_crossentropy(y_true, y_pred, from_logits=True)

def create_model(vocab_size, embed_dim, rnn_neurons, batch_size):
    model = Sequential()
    model.add(Embedding(vocab_size, embed_dim, batch_input_shape=[batch_size, None])) #vocab_size - input dimension
    model.add(GRU(rnn_neurons, return_sequences=True, stateful=True, recurrent_initializer='glorot_uniform'))
    model.add(Dense(vocab_size)) #output
    model.compile(optimizer='adam', loss=sparse_cat_loss) 
    return model

model = create_model(
  vocab_size = vocab_size,
  embed_dim=embed_dim,
  rnn_neurons=rnn_neurons,
  batch_size=batch_size)
model.summary()
```

```python
# Train the Model
#epochs = 30
#model.fit(dataset,epochs=epochs)
model = create_model(vocab_size, embed_dim, rnn_neurons, batch_size=1)
model.load_weights('shakespeare_gen.h5')
model.build(tf.TensorShape([1, None]))
model.summary()
```

```python
# Generating Text
# start_seed - Intial Seed text in string form
# gen_size - Number of characters to generate
# seed text -> format to input into network -> loop the sequence -> add predicted characters
def generate_text(model, start_seed, gen_size=100, temp=1.0): 
    num_generate = gen_size # Number of characters to generate
    input_eval = [char_to_ind[s] for s in start_seed] # Vecotrizing starting seed text
    input_eval = tf.expand_dims(input_eval, 0) # Expand to match batch format shape
    text_generated = [] # Empty list to hold resulting generated text
    
    # temperature - affects randomness - effect probability of next characters
    # Higher probability == lesss surprising/ more expected
    # Lower temperature == more surprising / less expected
    temperature = temp
    #batch size == 1
    model.reset_states()
    
    for i in range(num_generate):
        predictions = model(input_eval) # Generate Predictions
        predictions = tf.squeeze(predictions, 0) # Remove the batch shape dimension
        predictions = predictions / temperature # Use a cateogircal disitribution to select the next character
        predictions = predictions / temperature
        predicted_id = tf.random.categorical(predictions, num_samples=1)[-1,0].numpy()
        input_eval = tf.expand_dims([predicted_id], 0) # Pass the predicted charracter for the next input
        text_generated.append(ind_to_char[predicted_id]) # Transform back to character letter
    return (start_seed + ''.join(text_generated))

print(generate_text(model,"flower",gen_size=1000))
```

## Section 12: Generative Adversarial Networks

```python
# two networks competing against each other to generate data
# noise -> generator (generate images) -> discriminator (generated images vs real data) 
# -> generator keeps updating until discriminator can no longer tell the differences
# problems: training resources, mode collapse, instability - DCGANs (Deep Convolutional GANs)
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
plt.imshow(X_train[0]);
```

```python
# filter - smaller dataset
only_zeros = X_train[y_train==0]
only_zeros.shape
```

```python
np.random.seed(42)
tf.random.set_seed(42)

# 784 -> 150 -> 100 -> 150 -> 784
discriminator = Sequential()
discriminator.add(Flatten(input_shape=[28,28]))
discriminator.add(Dense(150,activation='relu'))
discriminator.add(Dense(100,activation='relu'))
discriminator.add(Dense(1,activation="sigmoid")) #output layer - binary classification: image real or fake
discriminator.compile(loss="binary_crossentropy", optimizer="adam")

codings_size = 100

generator = Sequential()
generator.add(Dense(100, activation="relu", input_shape=[codings_size]))
generator.add(Dense(150,activation='relu'))
generator.add(Dense(28 * 28, activation="sigmoid")) 
generator.add(Reshape([28,28]))

GAN = Sequential([generator, discriminator])
discriminator.trainable = False
GAN.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
GAN.layers[0].summary() #generator
```

```python
GAN.layers[1].summary() #discriminator
```

```python
# training batches
batch_size = 32
my_data = only_zeros
dataset = tf.data.Dataset.from_tensor_slices(my_data).shuffle(buffer_size=1000)
dataset = dataset.batch(batch_size, drop_remainder=True).prefetch(1) #the remainder of 5923 / 32 are dropped
epochs = 1
```

```python
generator, discriminator = GAN.layers
for epoch in range(epochs):
    print(f"Currently on Epoch {epoch+1}")
    i = 0
    for X_batch in dataset:
        i = i + 1
        if i % 100 == 0: #100 batches
            print(f"\tCurrently on batch number {i} of {len(my_data)//batch_size}")
        # train discriminator
        noise = tf.random.normal(shape=[batch_size, codings_size]) #create random noise
        gen_images = generator(noise) #Generate numbers based just on noise input
        #concatenate generated & real images - match data type
        X_fake_vs_real = tf.concat([gen_images, tf.dtypes.cast(X_batch,tf.float32)], axis=0)
        y1 = tf.constant([[0.]] * batch_size + [[1.]] * batch_size) #0 - fake, 1 - real
        discriminator.trainable = True # This gets rid of a Keras warning
        discriminator.train_on_batch(X_fake_vs_real, y1) # Train the discriminator on this batch
        
        # train generator
        noise = tf.random.normal(shape=[batch_size, codings_size])
        y2 = tf.constant([[1.]] * batch_size) # We want discriminator to belive that fake images are real
        discriminator.trainable = False
        
        GAN.train_on_batch(noise, y2)     
print("TRAINING COMPLETE")   
```
